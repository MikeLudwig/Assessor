```python
import numpy as np
import pandas as pd

work_dir = 'C:/Users/382365/Downloads/'
filename = 'Average_monthly_residential_energy_usage_By_zip_code.csv'
energydf = pd.read_csv(work_dir + filename)
```

```python
energydf.head()
```

```python
energydf = energydf.rename(columns={'?FY 05/06': 'FY 05/06', 'Lat': 'Latitude', 'Long': 'Longitude'})
```

```python
energydf
```

```python
energydf = energydf.drop(energydf.columns[[0]], axis=1)
```

```python
newcols = pd.DataFrame(energydf['Location 1'].str.split('\n',1).tolist(),columns = ['Zip','LatLong'])
```

```python
energydf = pd.concat([energydf, newcols], axis=1)
```

```python
energydf = energydf.drop('Location 1', 1)
```

```python
energydf
newcols = pd.DataFrame(energydf['LatLong'].str.split(',',1).tolist(),columns = ['Lat','Long'])
```

```python
newcols = newcols.apply(lambda x: x.str.strip('()'))
```

```python
energydf = pd.concat([energydf, newcols], axis=1)
```

```python
energydf = energydf.drop('LatLong', 1)
```

```python
energydf.to_csv(work_dir + filename)
```
